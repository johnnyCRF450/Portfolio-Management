```python
# Safest modeling
DAILY = get_returns(
                    stocks = ["MDLZ","PFE","NIO","UAL", "AMZN"],
                    wts =[0.1199,0.5906,0.0000,0.2338,0.0557],
                    start_date="2018-12-31")
DAILY.head()

```

```python
daily_returns = pyfolio.timeseries.cum_returns(
    returns= DAILY, starting_value=40000)

daily_returns
```

```python
from plotly.express import *

PLOT = area(x=daily_returns.index, y=daily_returns,
            color_discrete_sequence=["pink"])
PLOT.update_xaxes(rangeslider_visible=True)

PLOT.update_layout(template="plotly_dark",
                   font_family="Courier New")

PLOT
```
